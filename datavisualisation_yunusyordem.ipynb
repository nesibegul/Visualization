```python
def convert(row):
    # convert country name to country code
    try:
        cn_code = pc.country_name_to_country_alpha2(row.Entity, cn_name_format="default")
    except:
        cn_code = "Unknown"
    # convert cn_code to continent code
    try:
        conti_code = pc.country_alpha2_to_continent_code(cn_code)
    except:
        conti_code = "Unknown"
    return conti_code
# Yukarıdaki fonskiyon ülke isimlerinden kıta kısaltmaları için yeni bir değişken oluşturulmasını sağlar.
```

```python
df1["Continent"] = df1.apply(convert, axis=1)
df1.shape
df1.head()
```

```python
df1 = df1[df1.Continent != "Unknown"]
df1.shape
df1.head()
```

```python
df1["Continent"].unique()
```

Afrika kıtasında 2000, 2005, 2010 ve 2015 yıllarındaki sıtmanın görülme sıklığının karşılaştırılması<br>
Soru 4: Afrika kıtasında seneler süresince hastalığa yakalananlar arasındaki ölüm oranında bir ilişki var mı?

```python
df3_africa = df3[df3["Continent"] == "Africa"]
df3_africa.head()
df3_africa1 = df3_africa.copy()
```

```python
df3_africa1["Incidence of malaria (per 100,000 population at risk)"] =df3_africa1["Incidence of malaria (per 1,000 population at risk) (per 1,000 population at risk)"].apply(lambda x: x*100)
# Yukarıdaki işlem ile 1.000 kişide sıtmanın görülme sıklığı, 100.000 kişide görülme sıklığına çevrilmiştir.
```

```python
df3_africa1.head()
df3_africa1_year = df3_africa1.groupby("Year").agg({"Incidence of malaria (per 100,000 population at risk)": "mean"}).reset_index()
```

```python
df3_africa1_year.head()
```

```python
df1.head()
df1_africa = df1[df1["Continent"] == "Africa"]
df1_africa.head()
df1_africa_selected = df1_africa[df1_africa["Year"].isin([2000,2005,2010,2015])]
df1_africa_selected.head()
df1_africa_selected_year = df1_africa_selected.groupby("Year").agg({"Deaths - Malaria - Sex: Both - Age: Age-standardized (Rate) (per 100,000 people)": "sum"}).reset_index()
df1_africa_selected_year.head()
type(df1_africa_selected_year)
type(df3_africa1_year)
```

```python
merge_africa = pd.merge(df3_africa1_year, df1_africa_selected_year, on="Year")
merge_africa.head()
merge_africa["rate of those who die from disease"] = merge_africa["Deaths - Malaria - Sex: Both - Age: Age-standardized (Rate) (per 100,000 people)"]/merge_africa["Incidence of malaria (per 100,000 population at risk)"]*100
# Yukarıdaki işlem ile sıtmaya yakalanan kişilerden kaç kişinin hayatını kaybettiği bilgisi ortaya konmuştur.
```

```python
merge_africa.head()
```

Görselleştirme 4

```python
labels = "2000", "2005", "2010", "2015"
sizes = [merge_africa["rate of those who die from disease"][0], merge_africa["rate of those who die from disease"][1],
         merge_africa["rate of those who die from disease"][2],merge_africa["rate of those who die from disease"][3]]
```

```python
figure(figsize=(12, 8), dpi=80)
fig1, ax1 = plt.subplots()
colors = ["red", "gray", "blue", "green"]
patches, texts, autotexts = ax1.pie(sizes, colors=colors, labels=labels, autopct='%1.2f%%', startangle=90)
for text in texts:
    text.set_color("black")
for autotext in autotexts:
    autotext.set_color('white')
ax1.axis("equal")
plt.title(label="Rate of who die from disease",
          loc="center",
          fontweight="bold",
          fontfamily="serif")
plt.savefig("DeathRate_from_Disease.png")
plt.show()
```

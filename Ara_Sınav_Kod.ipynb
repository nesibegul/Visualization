```python
overall = df.drop(["year","prop"], axis = 1).groupby(["name", "sex"]).sum().sort_values("n", ascending = False).head(20) #year ve prop sütunları silindi, name ve sex'e göre gruplandı, n değerleri toplanıp sıralama yapıldı.
overall = overall.reset_index()   #indexler resetlendi.
overall = overall.sort_values(by = "n")   #değerler n'e göre sıralandı.
%matplotlib inline
plt.rcParams["figure.figsize"] = (20,9)   #figure boyutu belirlendi.
overall.plot.barh(x = "name", y = "n", rot = 0, color = "thistle", fontsize = 15)    #figure özellikleri belirlenerek ekrana basıldı.

plt.title("1880 ve 2017 Yılları Arasında En Çok Kullanılan İsimler", fontsize = 20)  #başlığın adı ve font büyüklüğü belirlendi.
plt.ylabel("İsimler", fontsize = 20)    #y ekseninin adı ve font büyüklüğü belirlendi.
plt.xlabel("Sayı (milyon) ", fontsize = 20)   #x ekseninin adı ve font büyüklüğü belirlendi.
plt.show()    #çizim ekrana yansıtıldı.
```

```python
the_slient_generation = df[(df["year"] >= 1928) & (df["year"] <= 1945)] #Sessiz kuşak yılları filtrelendi.
the_slient_generation = the_slient_generation.drop(["year","prop"], axis = 1).groupby(["name", "sex"]).sum().sort_values("n", ascending = False).head(20)   #year ve prop kolonları silindi, name ve sex'e göre gruplandı, n değerlerinin toplamına göre büyükten küçüğe sıralandı.
the_slient_generation = the_slient_generation.reset_index() #index resetlendi.
the_slient_generation = the_slient_generation.sort_values(by = "n")   #n değerine göre sıralandı.
%matplotlib inline
plt.rcParams["figure.figsize"] = (20,9)   #figure boyutu belirlendi.
the_slient_generation.plot.barh(x = "name", y = "n", rot = 0, color = "silver", xlabel = "İsimler",fontsize = "large")   #figure özellikleri belirlenerek ekrana basıldı.

plt.title("Sessiz Kuşağın (1928-1945) En Çok Kullanılan İsimleri", fontsize = 20)  #başlığın adı ve font büyüklüğü belirlendi.
plt.ylabel("İsimler", fontsize = 20)    #y ekseninin adı ve font büyüklüğü belirlendi.
plt.xlabel("Sayı (milyon) ", fontsize = 20)   #x ekseninin adı ve font büyüklüğü belirlendi.
plt.show()    #çizim ekrana yansıtıldı.
```

```python
#Michael Jordan'ın kariyer yıllarında, "Jordan" isminin kullanım sıklığı incelenmiştir.
#Michael Jordan'ın kariyer yılları açık pembe renk ile ifade edilmiştir. "Jordan" ismi kullanım sayısı da line - chart ile gösterilmiştir

michael_jordan = df.drop(["sex", "prop"], axis = 1)[df["name"] == "Jordan"]   #sex ve prop kolonları silindi, isim kısmı Jordan'a göre filtrelendi.
michael_jordan = michael_jordan.groupby("year").sum()   #liste yıllara göre gruplanarak, n sayısının toplamı yazıldı.
michael_jordan = michael_jordan.reset_index()   #index resetlendi.
plt.xlabel("Yıl", fontsize = 25)    #X ekseni font boyutu değiştirildi.
plt.ylabel("Miktar",fontsize = 25)  #Y ekseni font boyutu değiştirildi.
plt.title("Michael Jordan'ın Kariyer Yılları Boyunca 'Jordan' İsminin Kullanım Sıklığı", fontsize = 25)  #başlık yazıldı.
plt.rc('xtick', labelsize=25) #label size değiştirildi.
plt.rc('ytick', labelsize=25) #label size değiştirildi.
plt.grid()    #grid eklendi.
ax = plt.axes()
ax.set_facecolor('gainsboro')   #arka plan rengi değiştirildi.
plt.plot(michael_jordan.year, michael_jordan.n, color = "black")    #line grafiği çizgi rengi siyah olarak çizildi.
plt.fill_between(michael_jordan.year, michael_jordan.n, color = "lightblue", alpha = .5)    #line grafiği altında kalan alan açık mavi renk ile boyandı.
plt.axvspan(1984,2003, color = "red", alpha = 0.1)    #Michael Jordan'ın kariyer yılları transparan kırmızı renk ile boyandı.
```

```python
#Erkek ve kız isim sayılarının birbirine göre durumu yıllar içerisinde incelenmiştir.

gender_alaysis_female = df.drop(["name", "prop"], axis = 1)   #dataframe'den name ve prop kolonları silindi.
gender_alaysis_female = gender_alaysis_female[gender_alaysis_female["sex"] == "F"]    #sex female olacak şekilde filtrelendi.
gender_alaysis_female = gender_alaysis_female.groupby(["year"]).sum()   #year'a göre gruplanarak n sayısı toplandı.
gender_alaysis_female = gender_alaysis_female.reset_index()   #index resetlendi.

gender_alaysis_male = df.drop(["name", "prop"], axis = 1)   #dataframe'den name ve prop kolonları silindi.
gender_alaysis_male = gender_alaysis_male[gender_alaysis_male["sex"] == "M"]    #sex female olacak şekilde filtrelendi.
gender_alaysis_male = gender_alaysis_male.groupby(["year"]).sum()   #year'a göre gruplanarak n sayısı toplandı.
gender_alaysis_male = gender_alaysis_male.reset_index()   #index resetlendi.

plt.xlabel("Yıl", fontsize = 25)    #X ekseni ismi ve fontu belirlendi.
plt.ylabel("Miktar",fontsize = 25)    #Y ekseni ismi ve fontu belirlendi.
plt.title("'Erkek ve Kız İsim Sayılarının Zamanla Değişimi", fontsize = 25)   #Çizimin başlığı belirlendi.
plt.grid()    #grid eklendi.
ax = plt.axes()
ax.set_facecolor('gainsboro')   #arka plan rengi eklendi.
plt.plot(gender_alaysis_male.year, gender_alaysis_male.n, color = "blue")   #erkek isim sayıları grafiği mavi renk ile line chart olarak çizildi.
plt.plot(gender_alaysis_male.year, gender_alaysis_female.n, color = "purple")   #kadın isim sayıları grafiği mor renk ile line chart olarak çizildi.
plt.fill_between(gender_alaysis_male.year, gender_alaysis_male.n, color = "lightblue", alpha = .5)    #erkeklerin grafiğinin altı açık mavi ile boyandı.
plt.fill_between(gender_alaysis_female.year, gender_alaysis_female.n, color = "plum", alpha = .5)   #kızların grafiğinin altı plum rengi ile boyandı.
```

```python
#dataset içerisindeki toplam kız ve erkek isimlerinin birbirlerine oranları incelenmiştir.

pie = df.drop(["name", "prop"], axis = 1)   #dataframe'den isim ve prop kolonları silindi.
pie = pie.drop(["year"], axis = 1).groupby("sex").sum()   #dataframe'den yıl kolonu silindi, sex'e göre gruplandı, n sayıları toplandı.
pie = pie.reset_index()   #index resetlendi.
colors = ["plum", "lightskyblue"]   #pie-chart için renkler belirlendi.
textprops = {"fontsize":25}   #font size belirlendi.
wedgeprops = {"linewidth": 4, 'width':1, "edgecolor":"k"}   #pie-chart'în dışının kalın siyah çizgi ile çizilmesi sağlandı.
plt.pie(pie.n,labels =  pie.sex, radius = 1, autopct="%0.2f%%", explode = [0,0.1], colors = colors,textprops =textprops,  wedgeprops = wedgeprops)     #özellikler belirlenerek grafik çizildi.
plt.legend()    #legend çizildi.
plt.show()    #grafik gösterildi.
```

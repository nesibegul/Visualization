**Coffee Rating Verisinin Kaynaktan Çekilmesi**

```python
df=pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-07-07/coffee_ratings.csv")
```

```python
#Veride sadece çalşılacak sutünlar seçilmiştir.
df= df.loc[:,['total_cup_points', 'species', 'country_of_origin', 'grading_date','processing_method', 'aroma', 'flavor', 'aftertaste',
       'acidity', 'body', 'balance', 'uniformity', 'clean_cup', 'sweetness',
       'cupper_points', 'unit_of_measurement', 'altitude_mean_meters']]
```

```python
print(df.shape)
df.isnull().sum()
```

```python
#df.to_excel("df.xlsx")
```

```python
#Country_Of_Origin'de bulunan 1 Null değeri veriden silinmiştir.
df.dropna(subset=["country_of_origin"],inplace=True)
```

```python
#Country_Of_Origin=Honduras için 0 olan nicel değerler, ilgili sutün median değerleri ile doldurulmuştur
df_honduras=df[df.country_of_origin == "Honduras"].total_cup_points.to_frame()
honduras_median= df_honduras[df_honduras["total_cup_points"]!= 0].median()
print(honduras_median)
df["total_cup_points"].replace(0, 81.67, inplace=True)
df["aroma"].replace(0, 7.5, inplace=True)
df["flavor"].replace(0, 7.5, inplace=True)
df["aftertaste"].replace(0, 7.25, inplace=True)
df["acidity"].replace(0, 7.42, inplace=True)
df["body"].replace(0, 7.42, inplace=True)
df["balance"].replace(0, 7.33, inplace=True)
df["uniformity"].replace(0, 10, inplace=True)
df["clean_cup"].replace(0, 10, inplace=True)
df["sweetness"].replace(0, 10, inplace=True)
df["cupper_points"].replace(0, 7.33, inplace=True)
```

```python
#df[df["country_of_origin"]=='Honduras']
```

```python
df.isnull().sum()
```

```python
#Processing method boş olan verilerin other ile doldurduk
df["processing_method"].fillna("Other", inplace=True)
df.isnull().sum()
```

```python
#Grading date tarih formatına çevrilmiş ve gün/ay/yıl olacak şekilde parçalanmıştır.
df.grading_date = df.grading_date.apply(pd.to_datetime)
df["grading_month"]=df.grading_date.dt.month
df["grading_year"]=df.grading_date.dt.year
df["grading_day"]=df.grading_date.dt.day
df["grading_day_name"]=df.grading_date.dt.day_name()
print(df.shape)
df.sample(3)
```

```python
df.dtypes
```

```python
#Country_of_origin verisindeki uzun ifadeler kılastıldı anlamsız veriler düzeltildi
df["country_of_origin"].replace("Tanzania, United Republic Of", "Tanzania",inplace=True)
df["country_of_origin"].replace("United States (Puerto Rico)", "Puerto Rico",inplace=True)
df["country_of_origin"].replace("Papua New Guinea", "Papua",inplace=True)
df["country_of_origin"].replace("Cote d?Ivoire", "Cote",inplace=True)
```

```python
df["country_of_origin"].value_counts()
```

```python
#altitude_mean_meters sutününda 229 adet boş veri vadır. Bu sutünü sadece bir grafikte kullanıcağımız için veri seti yeni bir data frame'e kopyalanarak işleme devam edilmiştir. 
dfh= df.copy() 
```

```python
print(dfh.shape)
dfh.isnull().sum()
```

```python
#dfh veri setinde null değeri olan satırlar silinmiştir.
dfh.dropna(how ="any", inplace=True)
dfh.shape
```

```python
#dfh.tail(50)
```

```python
#altitude_mean_meters verilerinde ölçü birimi ft olanlar metreye dönüştürülmüştür
dfh.loc[dfh["unit_of_measurement"]=="ft", "altitude_mean_meters"]= dfh["altitude_mean_meters"]*0.3048
dfh.loc[dfh["unit_of_measurement"]=="ft"]
```

```python
# altitude_mean_meters verisi incelendiğinde verilerde aykırı değerler olduğu tespit edilmiştir.
dfh.altitude_mean_meters.describe()
```

```python
#altitude_mean_meters verisinin aykırı değerleri stuün median değerleri ile değiştirilmiştir. 
print(dfh.shape)
dfh.loc[dfh["altitude_mean_meters"]<100, "altitude_mean_meters"]= 1245
dfh.loc[dfh["altitude_mean_meters"]>10000, "altitude_mean_meters"]= 1245
dfh.altitude_mean_meters.describe()
sns.displot(dfh["altitude_mean_meters"],bins=16, color ="black")
```

**Verinin Görselleştirmeye Özel Derlenmesi**

```python
#1-Kahve türüne test sayısı ve ortalama puan grafikleri için ön hazırlık
df_species=df.groupby("species").median()
df_species=df_species["total_cup_points"]
df_species = df_species.to_frame()
df_counts= df.species.value_counts()
df_counts=df_counts.astype(float).to_frame()
df_counts
df_species["toplam_veri"]= df_counts
df_species["index"]= ['1','2']
df_species["species"]= ['Arabica','Robusta']
df_species.set_index("index",inplace=True)
```

```python
#2a-Kahve Mensei'ne göre Test Sayı Grafiği için ön hazırlık
counrtyCounts=df["country_of_origin"].value_counts()
counrtyCounts
Country_Mean=df.groupby("country_of_origin").mean()
Country_Mean=Country_Mean["total_cup_points"]
Country_Mean = Country_Mean.to_frame()
Country_Mean["toplam_veri"]= counrtyCounts
Country_Mean.reset_index()
Country_Mean=Country_Mean.sort_values(by="toplam_veri", ascending =False)
CountryName= Country_Mean.index
Country_Mean["Ulke_Name"] =CountryName
Country_Mean["Ulke_Name"].replace("United States (Hawaii)", "Hawaii",inplace=True)
Country_Mean=Country_Mean.loc[:"Kenya", :	]
Country_Mean.sort_values(by="total_cup_points",ascending=False)
```

```python
#4- Kahve Karakteristik Özelliği Analizi Radar Grafiği İçin Ön Hazırlık
#Yapılan analiz sonrasında kahvenin karakteristik özelliklerinin kaliteye etki edebileceğine karar verilmiş ve en iyi puana sahip 4 ülke için analiz yapılmıştır.
radar = df[df['country_of_origin'].isin(['Ethiopia','Kenya','Uganda','Colombia'])]
radar = radar.groupby('country_of_origin')[['aroma', 'flavor', 'aftertaste','acidity','body','balance','uniformity','clean_cup','sweetness']].mean().reset_index()
radar = pd.melt(radar, id_vars=['country_of_origin'], var_name='kategori', value_name='puan', value_vars=['aroma', 'flavor', 'aftertaste','acidity','body','balance','uniformity','clean_cup','sweetness'],)
print(radar)
data2={'country_of_origin': ["Diger","Diger","Diger","Diger","Diger","Diger","Diger","Diger","Diger"],
       'kategori': ["aroma", "flavor","aftertaste","acidity","body","balance","uniformity","clean_cup","sweetness"],
       "puan" : [7.52,7.46,7.32,7.48,7.45,7.43,9.85,9.84,9.89]}
data2 = pd.DataFrame(data2)
radar2= pd.concat([radar,data2],ignore_index=True)
radar2.reset_index()
```

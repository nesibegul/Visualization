```python
#drop the  null row
data.dropna(inplace = True)
```

```python
#Distribution of the median salaries by histogram

fig = plt.figure(figsize=(10,6))
data["Median"].plot.hist(bins=30)
plt.show()
```

```python
median_bycategory = data['Median']
sns.set_context(context = 'paper', font_scale =3)
plt.figure(figsize=(20,8))
ax = sns.barplot(data = data, x="Major_category", y="Median", )
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_title('Median Salary')
ax.legend()
```

```python
sns.set_context(context = 'paper', font_scale =3)
#sns.set_theme(font_scale=2)
plt.figure(figsize=(20,14))
ax = sns.boxplot(data=data, x="Major_category", y="Median", )
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_title('Median Salary')
ax.legend()
```

```python
#median_bycategory
data.groupby(by = 'Major_category').mean()['Median']
```

```python
#median_bycategory
major_categories = data['Major_category'].unique()#Different majer cartegories
sns.set_context(context = 'paper', font_scale =3)
#sns.set_theme(font_scale=2)
plt.figure(figsize=(20,14))
sns.barplot(x = data.groupby(by = 'Major_category').mean()['Median'].index, y =  data.groupby(by = 'Major_category').mean()['Median'])
ax.set_xticklabels(data.groupby(by = 'Major_category').mean()['Median'], rotation = 90)
ax.set_title('Median Salary by Major Category')
ax.legend()
```

```python
data1 =data.copy()
data1['Employment_ratio'] = data['Employed']/data['Total']
data1.head()
```

```python
#Employment ratio

sns.set_context(context = 'paper', font_scale =3)
#sns.set_theme(font_scale=2)
plt.figure(figsize=(20,14))
ax = sns.barplot(data=data1, x="Major_category", y="Employment_ratio" )
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_title('Number of Employed People by Major')
ax.legend()
```

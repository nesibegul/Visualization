### 6. Ortalama median ve standard sapma maaşa göre en çok kazandıran ilk beş fakülte
* Ortalama maaşların fakülteye göre ortalaması alınır ve büyükten küçüğe sıralanır
- Gelir ortalaması ve standart sapmaları birleştirelim

```python
#Ortalamanın bulunması
category_sorted = data[['Major_category', 'Median']].groupby(['Major_category']).mean() \
.sort_values(by = 'Median', ascending = False)
category_sorted.rename({'Median':'Mean_Income'}, axis = 1, inplace = True)

#Standart sapmaların bulunması
category_std = data[['Major_category', 'Median']].groupby(['Major_category']).std() \
.sort_values(by = 'Median', ascending = False)
category_std.rename({'Median':'Std_Income'}, axis = 1, inplace = True)

category_mean_std = pd.concat([category_sorted, category_std], axis=1, join="inner")
category_mean_std
```

### 7. Ilk 5 fakülteye ait ortalama maaşlar
* Veri setinin hazırlanması
* Grafiğin çizilmesi

```python
#median_bycategory
data_median_cat = pd.DataFrame(data.groupby(by = 'Major_category').mean()).sort_values(by = "Median",ascending = False)
data_median_cat
```

```python
#Sadece ilk 5 fakülteyi barplot 
category_mean_std_ilk5 = category_mean_std.head(5)

sns.set_context(context = 'notebook', font_scale =1)
plt.figure(figsize=(12,6))
current_palette = sns.color_palette("colorblind")
ax = sns.barplot(data = category_mean_std_ilk5, x=category_mean_std_ilk5.index, y='Mean_Income',  palette =current_palette )
ax.set_xticklabels(ax.get_xticklabels(), rotation=25,  fontdict={'fontsize':12})
ax.set_title('Ortalama Maaşı En fazla Olan İlk 5 Fakültenin Ortalaması', fontdict={'fontsize':12})
ax.set_xlabel("Fakülte İsimleri", fontdict={'fontsize':12})
ax.set_ylabel("Ortalama Maaşlar",   fontdict={'fontsize':12})
```

### 8. Fakültelere işe alınma oranı

```python
#Employment ratio
data1 =data.copy()
data1['Employment_ratio'] = 1- data1['Unemployment_rate']
data1 = data1.groupby(['Major_category']).mean()
data1 = data1.reset_index()
current_palette = sns.color_palette("colorblind")
data1.sort_values(by = 'Employment_ratio', ascending = False, inplace = True)
sns.set_context(context = 'paper', font_scale =2)
#sns.set_theme(font_scale=2)
plt.figure(figsize=(12,10))
ax = sns.barplot(data=data1, x="Major_category", y="Employment_ratio", palette=current_palette )
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_title('Fakülteye Göre İşe Alınma Oranları')
ax.legend()
```

### 9. Mühendislik fakultesindeki departmanların maaş ortalamaları

* Departman isimlerinde Engineering kelimesinin çıkarılması
* Grafiğin çizilmesi

```python
#Muhendislik fakulte isimlerinden  Engineering kelimesinin grafikte gosterimi icin atilmasi
data_eng = data[data['Major_category']=='Engineering']
remove_eng = lambda x: x.replace('ENGINEERING', '')
data_eng['Major'] = data_eng['Major'].apply(remove_eng).str.strip()

```

```python
sns.set_context(context = 'paper', font_scale =2)
#sns.set_theme(font_scale=2)
plt.figure(figsize=(20,16))
current_palette = sns.color_palette("colorblind")
ax = sns.barplot(data=data_eng, x="Major", y="Median",palette=current_palette )
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_title('Mühendislik Departmanlarının Ortalama Maaşları')
ax.set_xlabel("Departmanlar")
ax.set_ylabel("Ortalama maaşlar")
ax.legend()
```

### 10. Ortalama maaşların 25 çeyreklik ve 75 çeyrekliklerle beraber gösterilmesi

```python
#2 Ortalama maaş 3Q

fig, ax = plt.subplots()
sns.set_context(context = 'paper', font_scale =1)


ax.scatter(data_eng.Major,data_eng.Median, label="Maaş")
ax.scatter(data_eng.Major,data_eng.P25th, label="QP25")
ax.scatter(data_eng.Major,data_eng.P75th, label="QP75")

ax.legend()
plt.xticks(rotation = 90)
plt.title("Mühendislik departmanlarına göre maaşlar")
```
